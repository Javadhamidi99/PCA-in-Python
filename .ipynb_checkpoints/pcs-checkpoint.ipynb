Import the libraries and dataset

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#load dataset into pandas datarame
df = pd.read_csv(url , names =['sepal length','sepal width','petal length','petal width','target'])
print(df)
```

Data Standardization 

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
#getting the features
x = df.loc[ : , features].values
# getting the target
y = df.loc[ : , ["target"]].values
x = StandardScaler().fit_transform(x)
print(x[:5])
```

PCA Projecting to 2D

```python
pca_data = PCA(n_components=2)
principal_components = pca_data.fit_transform(x)
principaldf = pd.DataFrame (data = principal_components
                            , columns=['principal component 1', 'principal component 2'])
print(principaldf)
```

```python
finaldf = pd.concat ([principaldf,df[["target"]]], axis=1)
print(finaldf)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel("Principal Component 1" , fontsize =15)
ax.set_ylabel("Principal Component 2", fontsize = 15)
ax.set_title ("2 Componrnt PCA" , fontsize=20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target , color in zip(targets , colors) :
    indicestokeep = finaldf["target"] == target
    ax.scatter(finaldf.loc[indicestokeep,"principal component 1"]
               ,finaldf.loc[indicestokeep , "principal component 2"]
              ,c = color
              , s = 50)
ax.legend(targets)
ax.grid()
```

2nd usage of PCA 

PCA to Speed-Up Machine Learning Algorithms

---------------

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
mnist = fetch_openml("mnist_784")
```

```python
train_img , test_img , train_lbl , test_lbl =  train_test_split(mnist.data, mnist.target , test_size=1/7.0, random_state=0)
```

```python
scaler = StandardScaler()

scaler.fit(train_img)

train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
pca = PCA(0.95)
pca.fit(train_img)
train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
logistgreg = LogisticRegression(solver="lbfgs", max_iter=1000)
logistgreg.fit(train_img,train_lbl)
```

```python
logistgreg.predict(test_img[0:10])
```

```python
logistgreg.score(test_img, test_lbl)
```
